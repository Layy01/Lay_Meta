Code Showing all types of inheritance

```python
# Single Inheritance

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

dog = Dog()
print("Single Inheritance : ")
dog.speak()
dog.bark()

# Multiple Inheritance

class Reader :
  def read(self):
    print("Reading a book")

class Writer :
  def write(self):
    print("Writing a book")

class Author(Reader,Writer):
  def create(self):
    print("Creating a book")

author = Author()
print("\nMultiple Inheritance : ")
author.read()
author.write()
author.create()

# Multilevel Inheritance
class Animal:
  def __init__(self, name):
    self.name = name

  def speak(self):
    print(f"{self.name} is making a sound")

class Dog(Animal):
  def speak(self):
    print(f"{self.name} is barking")

class Puppy(Dog):
  def cry(self):
    print("puppy is crying")

animal = Animal("Jack")
dog = Dog("Charlie")
puppy = Puppy("Bella")

print("\nMultilevel Inheritance")
animal.speak()
dog.speak()
puppy.cry()

# Hiearachical Inheritance

class Vehicle:
    def move(self):
        print("Vehicle moves")

class Car(Vehicle):
    def drive(self):
        print("Car drives")

class Bike(Vehicle):
    def ride(self):
        print("Bike rides")

car = Car()
bike = Bike()

print("\nHiearachical Inheritance")
car.move()
car.drive()
bike.move()
bike.ride()

# Hybrid Inheritance
class Animal:
    def __init__(self, species):
        self.species = species

    def display_species(self):
        print(f"Species: {self.species}")

class Land(Animal):
    def move_on_land(self):
        print(f"{self.species} moves on land.")

class Water(Animal):
    def swim(self):
        print(f"{self.species} swims in the water.")

class Amphibian(Land, Water):
    def __init__(self, species):
        super().__init__(species)

    def display_info(self):
        print(f"{self.species} is an amphibian.")

frog = Amphibian("Frog")

print("\nHybrid Inheritance : ")
frog.display_info()
frog.display_species()
frog.move_on_land()
frog.swim()
```

Method Overloading

```python
class Addition:
    def add(self, a, b=0):
        return a + b

addition = Addition()

print(addition.add(10))
print(addition.add(200, 100))
```

Method Overriding

```python
class Animal:
    def sound(self):
        print("Animal makes a sound.")

class Dog(Animal):
    def sound(self):
        print("Dog barks.")

animal = Animal()
dog = Dog()

animal.sound()
dog.sound()
```
